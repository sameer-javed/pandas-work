# visualization in pandas  using matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:/Users/reals/Downloads/Ice Cream Ratings.csv")
df = df.set_index('Date') # Adds date as the main column 
df
```

```python
print(plt.style.available) # shows all tthe style available in matplotlib
```

```python
plt.style.use('fivethirtyeight')
# df.plot() # By Default it provides line plot of the data 
# We can add kind = line or any other chart name we want to add | the title will name the chart
# ylabel will assign the given name to y axis of the chart and xlabel will assign the given name to x axis
df.plot(kind='line' , title='Ice Cream ratings', ylabel='Daily ratings', xlabel='Score', figsize=(5,5)) 
# the subplot = True will show's diffrent line charts for each column
# df.plot(kind='line' , title='Ice Cream ratings', subplots=True)
```

```python
plt.style.use('seaborn-v0_8-deep') # changes the color of the graph
# Creates a bar plot 
df.plot(kind='bar',figsize=(5,5))
# Creates a stacked bar plot 
df.plot(kind='bar', stacked=True,figsize=(5,5)) 
# Creates a horizontal bar plot
df.plot.barh(figsize=(5,5))
```

```python

# creating a scatter plot 
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 100, c = 'red')
# s = 100 will increase the size of the dots 
# c = 'red' will change the color of the dots to red
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
# Creates the histogram diagram with defaullt size of bin = 10
# we can also increase the size of bins of decrease it as we want  
df.plot.hist(bins = 20)
```

```python
# creates boxplot
df.boxplot()
```

```python
plt.style.use('seaborn-v0_8-poster')
#creates area plot we can also increase it size by doing figsize = (10,5) it increase the size of the graph by 10 by 5
df.plot.area(figsize = (10,5))
```

```python
#creating pie chart remember it will always takes any one column to specify data of it 
df.plot.pie(y = 'Flavor Rating', figsize=(10,5))
```
